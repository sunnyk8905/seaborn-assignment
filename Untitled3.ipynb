```python
Line Plot:

Uses: Line plots are used to visualize the trend or relationship between two variables over a continuous axis, such as time. 
```

```python
Scatter Plot:

Uses: Scatter plots are useful for visualizing the relationship between two continuous variables. 
```

```python
Bar Plot:

Uses: Bar plots are ideal for comparing categorical variables or aggregating data based on categories. 
```

```python
Histogram:

Uses: Histograms are used to visualize the distribution of a single variable. 
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))


sns.boxplot(ax=axes[0], x='pclass', y='age', data=titanic_data)
axes[0].set_title('Age')


sns.boxplot(ax=axes[1], x='pclass', y='fare', data=titanic_data)
axes[1].set_title('Fare')


plt.tight_layout()


plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


diamonds_data = sns.load_dataset("diamonds")


sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)


plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Price with Cut')

plt.show()
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
```

```python
import seaborn as sns

iris_data = sns.load_dataset("iris")

sns.pairplot(data=iris_data, hue='species')


plt.show()
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


flights_data = sns.load_dataset("flights")


flights_matrix = flights_data.pivot("month", "year", "passengers")


sns.heatmap(data=flights_matrix, cmap="YlGnBu")


plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Flights Heatmap")


plt.show()
```
